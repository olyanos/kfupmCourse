## Q1

```python
df_ex = pd.read_csv('Data files/stud_exams.csv')
```

```python
df_ex
```

```python
mapper = {'completed':'yes', 'none': 'no'}
df_ex['Training'] = df_ex['test preparation course'].map(mapper)
```

```python
df_ex
```

```python
le = LabelEncoder()
df_ex["encoded_gender"] = le.fit_transform(df_ex['gender'])
```

```python
df_ex
```

```python
equal_w_binning = pd.DataFrame()
equal_w_binning['Math'] = pd.cut(df_ex['math score'],[0,35,75,100],labels=['low','mid','high'])
equal_w_binning['Reading'] = pd.cut(df_ex['reading score'],[0,35,75,100],labels=['low','mid','high'])
equal_w_binning['Writing'] = pd.cut(df_ex['writing score'],[0,35,75,100],labels=['low','mid','high'])
```

```python
# from chatGPT one liner solution
equal_w_binning['consistant'] = equal_w_binning.apply(lambda row: row['Math'] == row['Reading'] == row['Writing'], axis=1)
```

```python
equal_w_binning
```

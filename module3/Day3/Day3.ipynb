```python
df = pd.read_csv('Files/Germany.csv',index_col=0,parse_dates=True)
```

```python
df['Year'] = df.index.year
df['Month'] = df.index.month
df['Weekday'] = df.index.day_name()
```

```python
cols_to_plot = ['Consumption','Solar','Wind']
axes = df[cols_to_plot].plot(subplots=True,figsize=(10,10))
for ax in axes:
    ax.set_ylabel('Daily Totals (GwH)')
```

```python
ax = df.loc['2016-01':'2016-02','Consumption'].plot()
```

```python
import seaborn as sns
fig, axes = plt.subplots(3,1,figsize = (15,15),sharex=True)
for name,ax in zip(['Consumption','Solar','Wind'],axes):
    sns.boxplot(data=df,x='Month',y=name,ax=ax,palette='Set2')
    ax.set_ylabel('GWh')
    ax.set_title(name)
```

```python
import seaborn as sns
fig, axes = plt.subplots(3,1,figsize = (15,15),sharex=True)
for name,ax in zip(['Consumption','Solar','Wind'],axes):
    sns.boxplot(data=df,x='Weekday',y=name,ax=ax,palette='Set2')
    ax.set_ylabel('GWh')
    ax.set_title(name)
```
